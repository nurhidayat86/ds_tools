# Input necessary package

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Input data

```python
data = pd.read_csv("D:\Download\TONGDUN_FEAT_STATIC.csv")
```

# Metadata definition

```python
col_idex = 'SKP_CREDIT_CASE'
col_obs = 'FPD30_OBS'
col_target = 'FPD30_CNT'
col_time = 'DATE_DECISION'
col_month = 'MONTH'
col_day = 'DAY'
col_index = 'SKP_CREDIT_CASE'
```

```python
data['DATE_DECISION'] = pd.to_datetime(data['DATE_DECISION'], format='%Y-%m-%d')
```

```python
data[col_month] = data[col_time].dt.strftime('%Y%m').astype('int')
data[col_day] = data[col_time].dt.strftime('%Y%m%d').astype('int')
```
